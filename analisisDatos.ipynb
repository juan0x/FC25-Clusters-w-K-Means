```python
##Atributos Matriz correlación
corr_matrix = datos[["OVR","PAC","SHO","PAS","DRI","DEF","PHY"]].corr()#Extraer:Llamar el dataframe y el nombre de la columna o columnas como una lista

##Gráfica
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="winter", linewidths=0.5)

## El mapa de calor muestra las relaciones entre las estadíticas.
## Los valores positivos indican proporcionalidad Directa.
## Los valores negativos indican proporcionalidad Inversa.
```

```python
import warnings
warnings.filterwarnings("ignore")


attributes = ['PAC', 'SHO', 'DRI', 'PHY', 'DEF', 'PAS']
# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 15), sharey=True,)
fig.suptitle('Player attributes by position', fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot the distribution of each attribute in a separate subplot
for i, attr in enumerate(attributes):
    order = datos.groupby('Position')[attr].median().sort_values().index

    # Plot the boxplot with ordered categories
    sns.boxplot(x='Position', y=attr, data=datos, ax=axes[i], order=order, palette = 'Wistia')
    axes[i].set_title(f'{attr} rating by position')
    axes[i].set_ylabel(f'{attr}')
    axes[i].set_xlabel('Position')
# Adjust the layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
#Verificar donde no hay datos (son n/a o NaN)
hcalculo.isna().sum()
```

```python
##Imputar datos faltantes cambiando NaN por 0
### Esto no se debe hacer generalmente, pero en este caso no hay problema
gk_columns = ["GK Diving","GK Handling","GK Kicking","GK Positioning","GK Reflexes",]
hcalculo[gk_columns] = hcalculo[gk_columns].fillna(0)
```

```python
# Ahora extraeremos los datos
x = hcalculo.iloc[:,1:-1].values # "":"" = de todas las filas , 1 = la segunda: -1 = hasta la penultima. "":"" no incluye -1
y = hcalculo.iloc[:,-1].values # "":"" = de todas las columnas, -1 = hasta la última
y
```

```python
x
```

```python
#Numeralizar los valores categóricos (no numéricos): pie preferido, nacionalidad
##Codificar el pie preferido
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() ### Esto es una CLASE. Las clases van en primera letra mayúscula. Evitar la mala práctica de poner otras variables en mayúscula.

#Codificar penúltima columna (pie preferido)
x[:, -2]= le.fit_transform(x[:,-2])

#Codificar última columna (nacionalidad)
x[:, -1]= le.fit_transform(x[:,-1])

#Fit verifica cuantas entradas difetermtes en la columna y las codifica
#Transform cambia los valores en las columnas por su contraparte codificada

print(x[:,-2:]) # Se imprimen las últimas dos columnas para verificar la codificación
```

```python
datframe = pd.DataFrame(x) # Crear un dataframe a partir de pandas
for col in datframe.columns:
  datframe[col] = pd.to_numeric(datframe[col], errors = "coerce") #cambiar cada columna por un valor numérico
```

```python
x = datframe.values
x
```

```python
#Crear in imputador
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy = "median")#Reemplazar los valores nulos con ayuda de la mediana
imputer.fit(x)
x = imputer.transform(x)
x
```

```python
x_ = x[:,:-1] ## Eligiendo todos los datos excepto la nacionalidad
x_
```

```python
datos =  pd.DataFrame(x_)
datos.to_csv('datosbase_FC25.csv', index=False)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42) #se crea un objeto kmeans con las propiedades del algoritmo
clusters = kmeans.fit_predict(x_) # Asingando los datos y otorgandoles etiquetas (4, por número de clusters)
##Investigar método del codo
```

```python
# Ver centroides
centroides_kmeans = kmeans.cluster_centers_
centroides_kmeans
```

```python
#Añadir los clusters al Dataframe original
hcalculo["Cluster"]=clusters

print(hcalculo[["Name","Nation","Cluster"]].head())
```

#**Visualización**

```python
from sklearn.decomposition import PCA
#Crear un objeto PCA (Principal Component Analysis) y ajustar los datos
pca = PCA(n_components=3) #Queremos las tres primeras componentes principales
x_pca = pca.fit_transform(x_)
```

```python
varianza = pca.explained_variance_ratio_
print(sum(varianza))
```

```python
#crear un gráfico de dispersión 2D usando las dos primeras componentes
plt.figure(figsize=(10,6))
sns.scatterplot(x = x_pca[:,0], y = x_pca[:,1], hue=clusters,
                palette='viridis', alpha=0.6, s=100)
plt.title('Visualización 2D de la Clusterización (PCA)', fontsize=16)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns

def graficar_clusters_PCA3D(clusters, metodo):
    # Crear el gráfico 3D
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Graficar los puntos con borde blanco
    scatter = ax.scatter(x_pca[:, 0], x_pca[:, 1], x_pca[:, 2],
                         c=clusters, cmap='viridis', s=100,
                         edgecolor='white', alpha=0.6)
# Etiquetas de los ejes y título
    ax.set_title(f'Visualización 3D de la Clusterización {metodo}', fontsize=16)
    ax.set_xlabel('Componente Principal 1')
    ax.set_ylabel('Componente Principal 2')
    ax.set_zlabel('Componente Principal 3')

    # Añadir la leyenda
    legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
    ax.add_artist(legend1)

    # Mostrar el gráfico
    plt.show()
graficar_clusters_PCA3D(clusters, metodo='(PCA)')
```

```python
from sklearn.cluster import KMeans
## Inicializo donde almaceno el WCCS(k)
WCSS = []

for k in range(1,11):
  kpp = KMeans(n_clusters= k,
               #init = 'k-means++',
               random_state=42, n_init = 'auto')
  kpp.fit(x_)
  WCSS.append(kpp.inertia_)

plt.figure()
plt.plot(range(1,11), WCSS, 'o--', color = 'teal')
plt.title('Método del codo K means')
plt.xlabel('k')
plt.ylabel('WCSS')
plt.grid(True)
plt.grid(linestyle = '--')
plt.show()
```

El método del codo consiste en asignar el número de clusters en el punto el cual a a partir la distancia comienza a reducirse en menor proporción.
